```python
gpt_words_preproc_dict = {1: set(), 2: set(), 3: set()}

for key in gpt_words:
    for word in gpt_words[key]:
        # preprocess: stop-word removal, lemmatization
        doc = nlp(word)
        tokens = []
        for token in doc:
            if not (token.is_stop or token.is_punct or token.is_space):
                tokens.append(token.lemma_.lower())
                
                # also, if the unigram isn't already in our vocab, add it
                gpt_words_preproc_dict[1].add(token.lemma_.lower())
            
            text = " ".join(tokens)
            if len(tokens) in (2, 3):
                gpt_words_preproc_dict[len(tokens)].add(text)

for key in gpt_words_preproc_dict:
    print(len(gpt_words_preproc_dict[key]))
```
